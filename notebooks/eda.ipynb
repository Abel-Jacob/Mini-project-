```python
import os
import matplotlib.pyplot as plt
from PIL import Image

BASE_DIR = "../processed_data/dataset/train"

for label in ["genuine", "non_genuine"]:
    count = len(os.listdir(os.path.join(BASE_DIR, label)))
    print(f"{label}: {count} images")

sample_dir = os.path.join(BASE_DIR, "genuine")
sample_images = os.listdir(sample_dir)[:5]

plt.figure(figsize=(10,5))
for i, img_name in enumerate(sample_images):
    img = Image.open(os.path.join(sample_dir, img_name))
    plt.subplot(1,5,i+1)
    plt.imshow(img)
    plt.axis("off")
plt.show()
```
