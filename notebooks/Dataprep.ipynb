```python
import os
import shutil
import random

RAW_DATA_DIR = "../raw_data"
OUTPUT_DIR = "../processed_data/dataset"
SPLIT_RATIO = (0.7, 0.15, 0.15)
IMG_EXT = (".jpg", ".jpeg", ".png")

def create_dirs():
    for split in ["train", "val", "test"]:
        for label in ["genuine", "non_genuine"]:
            os.makedirs(os.path.join(OUTPUT_DIR, split, label), exist_ok=True)

def split_and_copy(brand_folder):
    images = [f for f in os.listdir(brand_folder) if f.lower().endswith(IMG_EXT)]
    random.shuffle(images)

    total = len(images)
    t1 = int(total * SPLIT_RATIO[0])
    t2 = t1 + int(total * SPLIT_RATIO[1])

    splits = {
        "train": images[:t1],
        "val": images[t1:t2],
        "test": images[t2:]
    }

    brand = os.path.basename(brand_folder).replace("_org", "")

    for split, files in splits.items():
        for i, file in enumerate(files):
            src = os.path.join(brand_folder, file)
            dst = os.path.join(
                OUTPUT_DIR, split, "genuine", f"{brand}_{i}.jpg"
            )
            shutil.copy(src, dst)

create_dirs()

for folder in os.listdir(RAW_DATA_DIR):
    if folder.endswith("_org"):
        split_and_copy(os.path.join(RAW_DATA_DIR, folder))

print("✅ Dataset prepared successfully")
```
